```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def linear_regression(x, y):
    # Remove missing or infinite values
    mask = np.isfinite(x) & np.isfinite(y)
    x = x[mask]
    y = y[mask]

    if len(x) == 0 or len(y) == 0:
        raise ValueError("No complete cases remaining after removing missing or infinite values.")

    # Calculate the coefficients (slope and intercept)
    slope = np.cov(x, y, ddof=0)[0, 1] / np.var(x, ddof=0)
    intercept = np.mean(y) - slope * np.mean(x)

    return intercept, slope

# Load dataset (example: car data)
data = pd.read_csv("/content/car data.csv")

# Select columns for linear regression
x = data['horsepower']
y = data['acceleration']

# Fit the linear regression model
intercept, slope = linear_regression(x, y)

# Print the coefficients
print("Intercept:", intercept)
print("Slope:", slope)

# Plot the data points
plt.figure()
plt.scatter(x, y, color='blue', label='Data')

# Add the regression line to the plot
plt.plot(x, slope*x + intercept, color='red', linewidth=2, label='Linear Regression')

plt.title('Linear Regression')
plt.xlabel('Horsepower')
plt.ylabel('Acceleration')
plt.legend()
plt.show()
```
